### Vizual interpretation

```python
# Show top 5 ATMs frequent at most

grouped_data = data.groupby('atmId')['numri_total_transactions'].sum()
grouped_data
```

```python
# Sort by number of transactions in descending order
sorted_data = grouped_data.sort_values(ascending=False)

# Get the top 5 ATMs
top_5_atms = sorted_data.head(20)
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
top_5_atms.plot(kind='bar', color='skyblue')
plt.title('Top 20 ATMs by Total Number of Transactions')
plt.xlabel('ATM ID')
plt.ylabel('Total Number of Transactions')
plt.xticks(rotation=40)
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.show()
```

```python
# Show most frequent day of the week to use ATMs


grouped_data = data.groupby('dita')['numri_total_transactions'].sum()
grouped_data
```

```python
x = grouped_data
y= grouped_data.index

plt.plot(y, x, linestyle = 'dashed', color = 'blue')
plt.title('Weekdays most frequent')
plt.xticks(rotation=30)
plt.savefig('Weekdays most frequent.png', format='png')
plt.show()
```

```python
# Show the balance between income and outcome number of transaction, do people deposit more or withdraw more money

data_out = data['numri_outgoing_transactions'].sum()
data_in = data['numri_incoming_transactions'].sum()

print(f"Outgoing transaction number: {data_out}, and incoming transaction number: {data_in}")
```

```python
# Data for pie chart
transaction_types = ['Outgoing', 'Incoming']
transaction_counts = [data_out, data_in]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(transaction_counts, labels=transaction_types, autopct='%1.1f%%', startangle=90, colors=['coral', 'skyblue'])
plt.title('Proportion of Outgoing and Incoming Transactions in total')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.savefig('Proportion of Outgoing and Incoming Transactions.png', format='png')

plt.show()
```

```python
### Show which days of the week is used more for transaction incoming vs outgoing

# Group by 'dita' and sum the transactions
grouped_outgoing = data.groupby('dita')['numri_outgoing_transactions'].sum()
grouped_incoming = data.groupby('dita')['numri_incoming_transactions'].sum()
```

```python
grouped_outgoing
```

```python
grouped_incoming
```

```python
# Ensure the days of the week are in the correct order
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
grouped_outgoing = grouped_outgoing.reindex(days_order)
grouped_incoming = grouped_incoming.reindex(days_order)
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(grouped_outgoing.index, grouped_outgoing.values, marker='o', linestyle='-', color='red', label='Outgoing Transactions')
plt.plot(grouped_incoming.index, grouped_incoming.values, marker='o', linestyle='-', color='blue', label='Incoming Transactions')

# Adding titles and labels
plt.title('Transactions Incoming vs Outgoing per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Transactions')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.savefig('Transactions Incoming vs Outgoing per Day of the Week.png', format='png')

# Display the plot
plt.show()
```
