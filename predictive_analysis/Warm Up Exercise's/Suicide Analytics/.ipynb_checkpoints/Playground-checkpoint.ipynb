```python
# Group by gender and sum the 'suicides_no' column
gender_suicides = df.groupby('sex')['suicides_no'].sum()
print(gender_suicides)

# Plot a bar chart
plt.figure(figsize=(14, 8))

# Create a bar chart using the index for the labels and the values for the heights
plt.bar(gender_suicides.index, gender_suicides.values, color= ['#ae2d68','#001e45'] )

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Total Suicides')
plt.title('Total Suicides by Gender', pad=20)
plt.savefig('Chart that shows suicide number based on gender.png', format='png')

# Show the plot
plt.show()
```
